```python
"""
This is the main program in a Jupyter Notebook file
"""


import pandas as pd
import datetime as dt
from screener import Screener

# initialize the dates using the datetime library
end_date = dt.date.today()
start_date = end_date-dt.timedelta(days=1*365)

# convert the dates for use in the screener
start_date_str = str(start_date)
end_date_str = str(end_date)

# load in the data using pandas
ticker_df = pd.read_csv("tickers.csv")
ticker_lst = ticker_df['Ticker'].tolist()

# initialize a screener object
my_screener = Screener(ticker_lst, start_date_str, end_date_str)

# retrieves the market data
my_screener.fetch_data()

# filters out the assets with high risk
my_screener.filter_assets(max_volatility=20)
my_screener.rank_assets()

# printing the final results using the generator
print("\nHere are the final recommended assets\n")
for safe_asset in my_screener.asset_generator():
    print(safe_asset)

# exports the data and plot the graph
my_screener.export_to_csv()
my_screener.plot_risk_vs_return()
```
